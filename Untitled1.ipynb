```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print('missing values in df \n',df.isnull().sum())
```

```python
#Dropping values
df_drop = df.dropna()
print('After deleting missing values\n',df_drop.isnull().sum())
```

```python
#fill with constants
df_fill_const = df.fillna("unknown")
print('After filling missing values\n',df_fill_const.head())
```

```python
#fill with numeric values
df_fill_mean = df.fillna(df.mean(numeric_only=True))
print('After filling missing values \n',df_fill_mean.head())
```
